```python
import cv2
import numpy as np
print("IMAGE PROCESSING PROJECT")
print("enter the location of the image or the image filename")
a="C:\\Users\\sairam\\Desktop\\download.jpeg" #make sure to give 2 slashes(\) in the address to avoid unicodeescape error
img=cv2.imread(a)

print("1.original 2.grayscale 3.RGB 4.blue 5.green 6.red 7.negative 8.grayscale_negative 9.canny edge 10.dilated edge")
print("11.eroded edge 12.autumn 13.bone 14.jet 15.winter 16.rainbow 17.ocean 18.summer 19.spring 20.cool ")
print("21.hsv 22.pink 23.hot 24.parula 25.magma 26.inferno 27.plasma 28.viridis 29.cividis 30.twilight")
print("31.twilight_shifted 32.turbo  33.deepgreen 34.flipped image 35.lab image")
print("enter the choice as number from (1-35)")
choice= int(input())
if choice==1:
    cv2.imshow('image',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==2:
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    cv2.imshow('grayscale',gray)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==3:
    rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    cv2.imshow('RGBimage',rgb)
    cv2.waitKey(0)
    cv2.destroyAllWindows() 
elif choice==4:
    b, g, r = cv2.split(img)
    zeros_ch = np.zeros(img.shape[0:2], dtype="uint8")
    blue_img = cv2.merge([b, zeros_ch, zeros_ch])
    cv2.imshow("Blue Image", blue_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows() 
elif choice==5:
    b, g, r = cv2.split(img)
    zeros_ch = np.zeros(img.shape[0:2], dtype="uint8")
    green_img = cv2.merge([ zeros_ch,g, zeros_ch])
    cv2.imshow("Green Image", green_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
elif choice==6:
    b, g, r = cv2.split(img)
    zeros_ch = np.zeros(img.shape[0:2], dtype="uint8")
    red_img = cv2.merge([zeros_ch, zeros_ch,r])
    cv2.imshow("Red Image", red_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()       
elif choice==7:
    neg=cv2.bitwise_not(img)
    cv2.imshow('Negative',neg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==8:
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    neg=cv2.bitwise_not(gray)
    cv2.imshow('Grayscale Negative',neg)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==9:
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(gray,100,150)
    cv2.imshow('Canny edge',canny)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==10:
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(gray,100,150)
    kernel=np.ones((3,3),np.uint8)
    imgdil= cv2.dilate(canny,kernel,iterations=1)
    cv2.imshow('Dilated image',imgdil)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
elif choice==11:
    gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    canny=cv2.Canny(gray,100,150)
    kernel=np.ones((1,1),np.uint8)
    imgero= cv2.erode(canny,kernel,iterations=1)
    cv2.imshow('Eroded image',imgero)
    cv2.waitKey(0)
    cv2.destroyAllWindows() 
elif choice==12:
    c=cv2.applyColorMap(img,0)
    cv2.imshow('Autumn colormap image',c)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==13:
    c=cv2.applyColorMap(img,1)
    cv2.imshow('Bone colormap image',c)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==14:
    c=cv2.applyColorMap(img,2)
    cv2.imshow('Jet colormap image',c)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==15:
    c1=cv2.applyColorMap(img,3)
    cv2.imshow('Winter colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==16:
    c1=cv2.applyColorMap(img,4)
    cv2.imshow('Rainbow colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==17:
    c1=cv2.applyColorMap(img,5)
    cv2.imshow('Ocean colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==18:
    c1=cv2.applyColorMap(img,6)
    cv2.imshow('Summer colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==19:
    c1=cv2.applyColorMap(img,7)
    cv2.imshow('Spring colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==20:
    c1=cv2.applyColorMap(img,8)
    cv2.imshow('Cool colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==21:
    c1=cv2.applyColorMap(img,9)
    cv2.imshow('HSV colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==22:
    c1=cv2.applyColorMap(img,10)
    cv2.imshow('Pink colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==23:
    c1=cv2.applyColorMap(img,11)
    cv2.imshow('Hot colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
elif choice==24:
    c1=cv2.applyColorMap(img,12)
    cv2.imshow('Parula colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
elif choice==25:
    c1=cv2.applyColorMap(img,13)
    cv2.imshow('Magma colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
elif choice==26:
    c1=cv2.applyColorMap(img,14)
    cv2.imshow('Inferno colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()   
elif choice==27:
    c1=cv2.applyColorMap(img,15)
    cv2.imshow('Plasma colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==28:
    c1=cv2.applyColorMap(img,16)
    cv2.imshow('Viridis colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==29:
    c1=cv2.applyColorMap(img,17)
    cv2.imshow('Civilis colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
elif choice==30:
    c1=cv2.applyColorMap(img,18)
    cv2.imshow('Twilight colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
elif choice==31:
    c1=cv2.applyColorMap(img,19)
    cv2.imshow('Twilight_shifted colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  
elif choice==32:
    c1=cv2.applyColorMap(img,20)
    cv2.imshow('Turbo colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==33:
    c1=cv2.applyColorMap(img,21)
    cv2.imshow('Deepgreen colormap image',c1)
    cv2.waitKey(0)
    cv2.destroyAllWindows() 
elif choice== 34:
    f=cv2.flip(img,0)
    cv2.imshow('Flipped image',f)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
elif choice==35:
    l=cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
    cv2.imshow('LAB image',l)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("Still not updated!")
    
print("THANK YOU FOR USING THIS! HOPE THIS HELPS!!")    
    
    
    
        
    
```
